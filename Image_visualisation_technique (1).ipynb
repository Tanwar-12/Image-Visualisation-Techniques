```python
# Image visualisation/Read images

image = mpimage.imread('/content/Screenshot (41).png')
plt.imshow(image)
plt.show()
```

```python
# Different different visualisation of image
# Isolate RGB image
r = image[:,:,0] # Red channel
g = image[:,:,1] # Green channel
b = image[:,:,2] # Blue channel

# Visualise the individual color channel
f, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r, cmap='Reds')
ax2.set_title('G channel')
ax2.imshow(g, cmap='Greens')
ax3.set_title('B channel')
ax3.imshow(b, cmap='Blues')

plt.show()
```

```python
# Convert greyscale for filtering
import matplotlib.image as mpimage

import cv2 
image= mpimage.imread('/content/Screenshot (41).png')
plt.imshow(image)
plt.show()
```

```python
# Converting to grey scale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray,cmap='gray')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray)
```

```python
import numpy as np
# Horizontal edge detection
sobel_y = np.array([[-1,-2,-1],
                    [0, 0, 0],
                    [1, 2, 1]])

# Verical edge detecion
sobel_x = np.array([[-1, 0 , 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

# Filter the image using filter2D(gray scal image, bit-depth kernel)
filter_image1 = cv2.filter2D(gray, -2, sobel_x)
filter_image2 = cv2.filter2D(gray, -2, sobel_y)

f, ax = plt.subplots(1,2, figsize=(15,4))
ax[0].set_title('Horizontal edge detection')
ax[0].imshow(filter_image1,cmap='gray')
ax[1].set_title('Vertical edge detection')
ax[1].imshow(filter_image2, cmap='gray')
```
